```python
import pandas as pd

# Step 1: Convert 'date' column to datetime format (outside the chunk loop for efficiency)
df_stores['state'] = df_stores['state'].astype(str)  # Ensure 'state' column is of type str

# Step 2: Get store_ids for the state 'Guayas'
store_ids = df_stores[df_stores['state'] == 'Guayas']['store_nbr'].unique()

# Step 3: Define the item families of interest
item_families = ['GROCERY I', 'BEVERAGES', 'CLEANING']

# Step 4: Initialize an empty list to hold filtered chunks
filtered_chunks = []

# Define chunk size (adjust based on system's memory capacity)
chunk_size = 10 ** 6

# Step 5: Read and filter the CSV file in chunks
for chunk in pd.read_csv('/content/drive/MyDrive/retail_kaggle_data/train.csv',
                         chunksize=chunk_size,
                         parse_dates=['date'],
                         dtype={'onpromotion': str}):
    # Filter by store_nbr, date range, and item families in the chunk
    chunk = chunk[(chunk['store_nbr'].isin(store_ids)) &
                  (chunk['date'] >= '2014-01-01') &
                  (chunk['date'] <= '2014-03-31')]

    # Merge with df_items (we assume df_items is already loaded)
    chunk = chunk.merge(df_items, on='item_nbr', how='left')

    # Filter by item families of interest
    chunk = chunk[chunk['family'].isin(item_families)]

    # Append the filtered chunk to the list
    filtered_chunks.append(chunk)

    # Free memory by deleting the chunk after processing
    del chunk

# Step 6: Concatenate all filtered chunks into a single DataFrame
df_train_filtered = pd.concat(filtered_chunks, ignore_index=True)

# Step 7: Save the filtered dataset
df_train_filtered.to_csv('/content/filtered_train_guayas_families.csv', index=False)

# Step 8: Print the number of rows after filtering
print(f"Data after filtering by item families: {df_train_filtered.shape[0]} rows")
```

## Step 3: Handle Outliers

```python
# Step 3: Handle Outliers
# This step identifies and removes extreme outliers using Z-score analysis.

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

def calculate_store_item_zscore(group):
    mean_sales = group['unit_sales'].mean()
    std_sales = group['unit_sales'].std()
    group = group.copy()  # Ensure modifications don't affect the original DataFrame
    group['z_score'] = (group['unit_sales'] - mean_sales) / (std_sales if std_sales != 0 else 1)
    return group

# Apply Z-score calculation to each store-item combination
df_filterd_train = df_filterd_train.groupby(['store_nbr', 'item_nbr'], group_keys=False).apply(calculate_store_item_zscore)

# Define threshold for outliers (e.g., Z-score > 5)
outliers = df_filterd_train[df_filterd_train['z_score'].abs() > 5]
print(f"Number of outliers detected: {len(outliers)}")

# Remove outliers
df_filterd_train = df_filterd_train[df_filterd_train['z_score'].abs() <= 5]

# Drop the Z-score column as it is no longer needed
df_filterd_train = df_filterd_train.drop(columns=['z_score'])

# Step 4: Save the cleaned dataset
# The to_csv() method only saves the dataframe to disk, it doesn't return it.
# So we need to assign df_train_filtered to df_train_cleaned before saving it.
df_train_cleaned = df_filterd_train
df_train_cleaned.to_csv('/content/cleaned_train_guayas.csv', index=False)
print("Cleaned dataset saved: /content/cleaned_train_guayas.csv")
```

```python
import dask.dataframe as dd
import pandas as pd
import numpy as np

# Ensure df_train_cleaned is a Dask DataFrame
if not isinstance(df_train_cleaned, dd.DataFrame):
    df_train_cleaned = dd.from_pandas(df_train_cleaned, npartitions=20)

# Convert 'date' to datetime format to avoid merge errors
df_train_cleaned['date'] = dd.to_datetime(df_train_cleaned['date'])

# Compute min and max date efficiently
min_date = df_train_cleaned['date'].min().compute()
max_date = df_train_cleaned['date'].max().compute()
full_date_range = pd.date_range(start=min_date, end=max_date, freq='D')

# Compute unique store-item combinations efficiently
store_item_combinations = df_train_cleaned[['store_nbr', 'item_nbr']].drop_duplicates().compute()

# **Step 1: Create only relevant date ranges for each store**
store_min_max_dates = df_train_cleaned.groupby('store_nbr')['date'].agg(['min', 'max']).compute()

# **Step 2: Generate only valid store-item-date combinations**
partitioned_combinations = []
for store, (start_date, end_date) in store_min_max_dates.iterrows():
    date_range = pd.date_range(start=start_date, end=end_date, freq='D')

    # Filter items only for this store
    store_items = store_item_combinations[store_item_combinations['store_nbr'] == store]

    partial_index = pd.DataFrame(
        [(date, store, item) for date in date_range for item in store_items['item_nbr']],
        columns=['date', 'store_nbr', 'item_nbr']
    )

    # Convert to Dask for efficient processing
    partitioned_combinations.append(dd.from_pandas(partial_index, npartitions=2))

# Concatenate all partitions
full_date_store_item = dd.concat(partitioned_combinations)

# Ensure 'date' is in datetime format before merging
full_date_store_item['date'] = dd.to_datetime(full_date_store_item['date'])

# **Step 3: Merge while keeping only matched records**
df_filled = full_date_store_item.merge(
    df_train_cleaned,
    on=['date', 'store_nbr', 'item_nbr'],
    how='left'
)

# **Step 4: Fill missing values and ensure proper data types**
df_filled['unit_sales'] = df_filled['unit_sales'].fillna(0).astype(float)
df_filled['onpromotion'] = df_filled['onpromotion'].fillna(False)

# Convert Dask DataFrame types using map_partitions
df_filled = df_filled.map_partitions(lambda df: df.astype({'onpromotion': bool}))

# Convert objects to proper types to prevent FutureWarnings
df_filled = df_filled.map_partitions(lambda df: df.infer_objects(copy=False))

# **Step 5: Ensure we only keep valid store-item-date rows**
df_filled = df_filled[df_filled['date'] >= df_train_cleaned['date'].min().compute()]

# Drop unnecessary columns (e.g., id if it is not needed)
df_filled = df_filled.drop(columns=['id'], errors='ignore')

# **Step 6: Sort and compute efficiently**
df_filled = df_filled.sort_values(by=['date', 'store_nbr', 'item_nbr']).compute()

# **Check final size**
print(f"Final dataset shape: {df_filled.shape}")

# Check the result
print(df_filled.head())
```

# Autocorrelation Function (ACF) Plot

```python
from statsmodels.graphics.tsaplots import plot_acf
import numpy as np

# Ensure no missing values
sales_by_date = df_filled.groupby('date')['unit_sales'].sum()
sales_array = np.array(sales_by_date)

# Autocorrelation Plot
plt.figure(figsize=(10, 5))
plot_acf(sales_array, lags=12 * 7)  # Check up to 12 weeks for weekly patterns
plt.title(' Autocorrelation of Sales Data (Weekly Patterns)', fontsize=18, fontweight='bold')
plt.grid(True)
plt.show()
```

This Autocorrelation Function (ACF) Plot helps identify patterns and seasonality in sales data over time.

Key Observations:

Strong initial autocorrelation (Lag 0):
This is expected since the correlation of a time series with itself at lag 0 is always 1.
Significant Positive Peaks at Regular Intervals:
Notice the higher peaks at around lag 7, 14, 21, etc. (multiples of 7).
This suggests a weekly seasonality, meaning sales tend to repeat similar patterns every week.
Gradual Decay Over Time:
As the lag increases, correlations gradually decrease.
This means that sales data from distant days become less related to the present day.
Alternating Positive and Negative Correlations:
Some lags show negative correlation, meaning that sales on those days are opposite (e.g., high sales followed by lower sales).
This might indicate demand fluctuations (e.g., weekend vs. weekday sales differences).
Shaded Confidence Interval:
The blue-shaded area represents the 95% confidence interval.
If points fall outside this region, they are considered statistically significant.
Here, multiple lags exceed the confidence bounds, confirming real periodicity in the data.
What This Means for Forecasting:

Since strong correlations appear every 7 days, our model should incorporate a weekly seasonal component.
Lags of 7, 14, and 21 days should be used as features in time series models (e.g., SARIMA, XGBoost, or LSTM).
Expanding and rolling averages should also be calculated for weekly windows to capture the pattern.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the oil data
df_oil = pd.read_csv('/content/oil.csv')
df_oil['date'] = pd.to_datetime(df_oil['date'])

# Set the 'date' column as the index
df_oil.set_index('date', inplace=True)

# Create a complete date range based on the date range in your sales data (assuming df_filled contains sales data)
full_date_range = pd.date_range(start=df_oil.index.min(), end=df_oil.index.max(), freq='D')

# Reindex the oil dataframe with the complete date range
df_oil_reindexed = df_oil.reindex(full_date_range)

# Fill missing oil prices using forward fill
df_oil_reindexed['dcoilwtico'] = df_oil_reindexed['dcoilwtico'].fillna(method='ffill')

# Merge the filled oil prices with your sales data (assuming df_filled contains the sales data)
df_train_oil = pd.merge(df_filled_items, df_oil_reindexed, left_on='date', right_index=True, how='left')

# Plotting oil price vs unit sales
fig, ax1 = plt.subplots(figsize=(10,6))

ax1.set_xlabel('Date')
ax1.set_ylabel('Oil Price', color='tab:blue')
ax1.plot(df_train_oil['date'], df_train_oil['dcoilwtico'], color='tab:blue', label='Oil Price')
ax2 = ax1.twinx()
ax2.set_ylabel('Unit Sales', color='tab:green')
ax2.plot(df_train_oil['date'], df_train_oil['unit_sales'], color='tab:green', label='Unit Sales')

plt.title('Oil Price vs Unit Sales Over Time', fontsize=16)
plt.show()
```

# ADF Test

```python
from statsmodels.tsa.stattools import adfuller

# Aggregate total sales by date
sales_by_date = df_filled.groupby('date')['unit_sales'].sum()

# Perform the Augmented Dickey-Fuller (ADF) test
adf_test = adfuller(sales_by_date)

# Print results
print(" Augmented Dickey-Fuller Test Results:")
print(f"ADF Statistic: {adf_test[0]:.4f}")
print(f"p-value: {adf_test[1]:.4f}")
print(f"Critical Values: {adf_test[4]}")

# Interpretation
if adf_test[1] < 0.05:
    print(" The data is stationary (Reject H)")
else:
    print(" The data is non-stationary (Fail to Reject H)")
```
